## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv("data\hist_data.csv")

fig, ax = plt.subplots()
ax.hist(df["aデータ"], bins=30, alpha=0.5, color="red", label="aデータ")
ax.hist(df["bデータ"], bins=30, alpha=0.5, color="turquoise", label="bデータ")

ax.set_title("aデータとbデータの温度に関する度数分布")
ax.set_xlabel("温度(℃)")
ax.set_ylabel("度数")
ax.legend()
#- 共通するもの: ヒストグラムの数: 30, 透明度: 0.5
#- 違うもの: 色(aデータ: 赤, bデータ: ターコイズブルー)
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df = pd.read_csv("data\scatter_data.csv")
fig, ax = plt.subplots()

colors = ["red", "blue", "green"]

for class_num in range(3):
    df_by_class = df[df["class"] == class_num]
    ax.scatter(
        df_by_class["length (cm)"],
        df_by_class["width (cm)"],
        color=colors[class_num],
        label=f"class{class_num}"
    )

ax.set_title("クラス0～2のlengthとwidthに関する散布図")
ax.set_xlabel("length (cm)")
ax.set_ylabel("width (cm)")
ax.grid()
ax.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
#- 問3の棒グラフ
#    - 横軸の文字を回転させる角度: 30
#- 問3の折れ線グラフ
#    - マーカーの形状: 円, 横軸の文字を回転させる角度: 30
df = pd.read_csv("data\line_data.csv")

#折れ線グラフ
fig1, ax1 = plt.subplots()

ax1.plot(df["期間"], df["件数"], marker="o")
ax1.tick_params(axis="x", rotation=30)
ax1.set_title("期間ごとの件数の推移")
ax1.set_xlabel("期間")
ax1.set_ylabel("件数")
ax1.grid()


#棒グラフ
fig2, ax2 = plt.subplots()

ax2.bar(df["期間"], df["費用"])
ax2.tick_params(axis="x", rotation=30)
ax2.set_title("期間ごとの費用の推移")
ax2.set_xlabel("期間")
ax2.set_ylabel("費用(円)")
ax2.grid()
```
